```python
import matplotlib_venn as venn
set1 = set(output_df["precursor"])
set2 = set(comparison_df["precursor"])

venn.venn2([set1, set2], set_labels = ("output_df", "comparison_df"))

output_df_reduced = output_df.drop(columns=["run"]).reset_index(drop=True).drop_duplicates()
comparison_df_reduced = comparison_df.drop(columns=["replicate"]).reset_index(drop=True).drop_duplicates()

output_df_intersection_set = set(output_df_reduced["ion"]).intersection(set(comparison_df_reduced["ion"]))

output_df_intersection = output_df_reduced[output_df_reduced["ion"].isin(output_df_intersection_set)]


print(output_df_intersection.equals(comparison_df_reduced))

display(output_df_intersection)
display(comparison_df_reduced)
```
