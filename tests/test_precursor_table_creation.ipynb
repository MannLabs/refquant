```python

%reload_ext autoreload
%autoreload 2


import refquant.reference_quantification.multi_run.multi_run_table_creation as multi_run_table_creation
import refquant.table_preparation.table_import as table_import
import refquant.reference_quantification.multi_precursor.loading_manager as loading_manager

test_input = "../test_data/diann_test_input.tsv"
reformatted_precursor_table = "../test_data/diann_test_input.tsv.diann_precursors_mDIA.aq_reformat.tsv"
```

```python
import pandas as pd
reference_table = table_import.TableReformatterDIANN(test_input).filename_reformatted
```

```python
def load_reformatted_precursor_df(reformatted_precursor_table):
    df_precursor_reformatted = pd.read_csv(reformatted_precursor_table, sep="\t")
    relevant_columns = [x for x in df_precursor_reformatted.columns if not x.endswith("_(Dimethyl-n-0)")]
    df_precursor_reformatted = df_precursor_reformatted[relevant_columns]
    renamed_columns = [x.replace("_(Dimethyl-n-4)", "_target4").replace("_(Dimethyl-n-8)", "_target8") for x in df_precursor_reformatted.columns]
    df_precursor_reformatted.columns = renamed_columns
    return df_precursor_reformatted
```

```python
df_precursor_reformatted = load_reformatted_precursor_df(reformatted_precursor_table)
```

```python
single_labelled_precursors = loading_manager.get_all_single_labelled_precursors_in_dataset_diann(reference_table, use_multiprocessing=True, number_of_cores=10)
```

```python
reference_df_after_proc = multi_run_table_creation.PrecursorTableCreatorQuantityBySearchEngine(single_labelled_precursors).precursorquantitytable
```
